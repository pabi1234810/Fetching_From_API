```python
import http.client
import json

conn = http.client.HTTPSConnection("free-api-live-football-data.p.rapidapi.com")

headers = {
    'x-rapidapi-key': "2ff1a5cdb1msh95170e496a27ed1p1ecd73jsn484bda694c7b",
    'x-rapidapi-host': "free-api-live-football-data.p.rapidapi.com"
}

# Send request
conn.request("GET", "/football-players-search?search=m", headers=headers)

# Get response
res = conn.getresponse()
data = res.read()

# Decode response
decoded_data = data.decode("utf-8")

# Convert to JSON (dictionary in Python)
json_data = json.loads(decoded_data)

# Print pretty JSON
print(json.dumps(json_data, indent=4))

# Save JSON to file
with open("football_players.json", "w", encoding="utf-8") as f:
    json.dump(json_data, f, indent=4, ensure_ascii=False)
```
